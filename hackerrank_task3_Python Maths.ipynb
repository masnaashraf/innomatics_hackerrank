**Q1-Polar Coordinates**


```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

import cmath
c=complex(input().strip())
for i in cmath.polar(c):
    print(i)
```

**Q2- Find Angle MBC**


```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import math
AB = int(input())
BC = int(input())
print(round(math.degrees(math.atan(AB/BC))),chr(176),sep='')
```

**Q3-mod divmod**


```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

a=int(input())
b=int(input())

print(a//b)
print(a%b)
print(divmod(a,b))
```
