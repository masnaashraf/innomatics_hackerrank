**Q4-Find a string**


```python
def count_substring(string, sub_string):
    str_len=len(string)
    sub_str_len=len(sub_string)
    count = 0
    
    for i in range(str_len-sub_str_len+1):
        if(string[i:(i+sub_str_len)]==sub_string):
           count=count+1
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

**Q7-Text Wrap**


```python
import textwrap

def wrap(string, max_width):
    chunks=[string[i:i+max_width] for i in range(0,len(string),max_width)]
    str_new="\n".join(map(str,chunks))
    return str_new

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

**Q12-The Minion Game**


```python
 def minion_game(string):
    # your code goes here
    length = len(string)
    stuart_score,kevin_score= 0,0
    for i in range(length):
        if(string[i] in 'AEIOU'):
            kevin_score=kevin_score+(length-i)
        else:
            stuart_score=stuart_score+(length-i)
    if(stuart_score>kevin_score):
        print("Stuart {}".format(stuart_score))
    elif(stuart_score==kevin_score):
        print("Draw")
    else:
        print("Kevin {}".format(kevin_score))

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

**q14-swap case**


```python
def swap_case(s):
    swapped = []
    for letter in s:
        if letter == letter.lower():
            swapped.append(letter.upper())
        elif letter == letter.upper():
            swapped.append(letter.lower())
    
    return ''.join(swapped)
    return

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```
